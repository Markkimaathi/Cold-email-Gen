```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This document is about Nairobi",
        "This document is about Kilimani"
    ],
    ids = ['id1', 'id2']
)
```

```python
all_docs = collection.get()
all_docs
```

```python
documents = collection.get(ids=["id1"])
documents
```

```python
results = collection.query(
    query_texts=['Query is about chapati'],
    n_results=2
)
results
```

```python
collection.delete(ids=all_docs['ids'])
collection.get()
```

```python
collection.add(
    documents=[
        "This document is about Kilimani",
        "This document is about Nairobi"
    ],
    ids=["id3", "id4"],
    metadatas=[
        {"url": "https://en.wikipedia.org/wiki/Kilimani"},
        {"url": "https://en.wikipedia.org/wiki/Nairobi"}
    ]
)
```

```python
results = collection.query(
    query_texts=["Query is about chapati"],
    n_results=2
)
results
```
